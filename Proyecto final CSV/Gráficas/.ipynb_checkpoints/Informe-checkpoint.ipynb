### 7. ¿Cuál es el equipo con mayor posibilidad de ganar el título en la temporada?

```python
import pandas as pd
import matplotlib.pyplot as plt
path_archivos = "/home/dev-16/Documentos/Universidad/Python-Revelo-Karla/py-revelo-herrera-karla-michelle/Proyecto final CSV/Archivos/"

playoff_prob_columns = ['Equipo', 'Probabilidad']
df = pd.read_csv(path_archivos + 'win_probs.csv', usecols = playoff_prob_columns);
df = df.sort_values(by ='Probabilidad' ,ascending=False)
df = df[df['Probabilidad']>5]
colors = ['red', 'gold']
fig, ax1 = plt.subplots()
ax1.pie(
    df['Probabilidad'], 
    labels = df['Equipo'],
    explode = [0.05] * len(df['Equipo']),
    autopct = '%1.2f%%', 
    shadow = True, 
    radius = 1,
    colors = colors,
    startangle = 0)
fig.suptitle('Equipo con mayor posibilidad de ganar el título', fontsize=20)
fig.figsize = (5,5)
plt.show()
```

### 19. ¿Cuál es la proyección del MVP del 2020 frente al del 2019?

```python
import pandas as pd
import matplotlib.pyplot as plt
from math import pi
path_archivos = "/home/dev-16/Documentos/Universidad/Python-Revelo-Karla/py-revelo-herrera-karla-michelle/Proyecto final CSV/Archivos/"

stats_columns = ['TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS']

df = pd.read_csv(path_archivos + 'mvp_stats.csv', usecols = stats_columns);
AttNo = len(stats_columns)
datos_2019 = df.iloc[0].tolist()
datos_2019 +=  datos_2019[:1]

datos_2020 = df.iloc[1].tolist()
datos_2020 += datos_2020[:1]

angles = [n / float(AttNo) * 2 * pi for n in range(AttNo)]
angles += angles [:1]

angles2 = [n / float(AttNo) * 2 * pi for n in range(AttNo)]
angles2 += angles2 [:1]

ax = plt.subplot(111, polar=True)

plt.xticks(angles,stats_columns)

ax.plot(angles,datos_2019)
ax.fill(angles, datos_2019, 'teal', alpha=0.1)


ax.plot(angles2, datos_2020)
ax.fill(angles2, datos_2020, 'red', alpha=0.1)

#Give the plot a title and show it
plt.figtext(0.2,0.9,"MVP 2020",color="black")
plt.figtext(0.2,0.86,"vs")
plt.figtext(0.2,0.8,"MVP 2019",color="black")
plt.show()
plt.show()
```
