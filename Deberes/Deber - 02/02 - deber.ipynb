```python
import numpy as np
import pygame

IMAGE_FILE = "Bob-Esponja - 2.jpg" 
IMAGE_SIZE = (660, 660)
r_vacia = 0
c_vacia = 0

def juego_rompecabezas():
    global r_vacia, c_vacia
    num = int(input('Ingrese la dimension que desee: '))    
    COLUMNS = num
    ROWS = num    
    movimientos = 20
    PIEZA_WIDTH = IMAGE_SIZE[0] / COLUMNS
    PIEZA_HEIGHT = IMAGE_SIZE[1] / ROWS
    
    if(num == 2):
        numero_random = 2
    elif (num == 3):
        numero_random = 3
    elif (num == 4):
        numero_random = 4
    elif (num == 5):
        numero_random = 5
    elif (num == 6):
        numero_random = 6
    else:
        numero_random = 7

    #Representar la imagen
    hor_border = pygame.Surface((int(PIEZA_WIDTH), 2))
    ver_border = pygame.Surface((2, int(PIEZA_HEIGHT)))

    #Añadir bordes a las piezas y dividir la imagen
    imagen = pygame.image.load(IMAGE_FILE)
    piezas = {}
    for r in range(ROWS):
        for c in range(COLUMNS):
            pieza = imagen.subsurface(int(r*PIEZA_HEIGHT), int(c*PIEZA_WIDTH), int(PIEZA_HEIGHT), int(PIEZA_WIDTH))
            pieza.blit(hor_border, (0, 0))
            pieza.blit(ver_border, (0, 0))
            piezas[(r,c)] = pieza


    #Rellenar pieza vacia
    pieza_vacia = (np.random.randint(numero_random),np.random.randint(numero_random))
    piezas[pieza_vacia].fill((0,0,0))
    (r_vacia, c_vacia) = pieza_vacia 

    #Posiciones
    posicion_inicial = {(row,col): (row,col) 
                      for row in range(ROWS) for col in range(COLUMNS)}
    posicion = {(row,col): (row,col) 
              for row in range(ROWS) for col in range(COLUMNS)}

    # Juego y rompecabezas completo
    pygame.init()
    display = pygame.display.set_mode((IMAGE_SIZE))
    pygame.display.set_caption("Rompecabezas")
    display.blit (imagen, (0, 0))
    pygame.display.flip() #Actualizar pantalla 



    def intercambiar_piezas(r,c):
         global c_vacia, r_vacia
         display.blit(piezas[posicion[(r, c)]], (int(r_vacia*PIEZA_HEIGHT), int(c_vacia*PIEZA_WIDTH))) #Dibujo la pieza vacia
         display.blit(piezas[pieza_vacia], (int(r*PIEZA_HEIGHT), int(c*PIEZA_WIDTH))) #Dibujo la pieza que quiero mover
         posicion[(r_vacia, c_vacia)] = posicion[(r, c)] #Cambio de posiciones
         posicion[(r, c)] = pieza_vacia
         (r_vacia, c_vacia) = (r, c)
         pygame.display.flip() #Actualizar imagen
         #print(posicion)


    def ordena_piezas():
        for i in range(movimientos):
            intercambiar_piezas(np.random.randint(numero_random),np.random.randint(numero_random))
            pygame.time.delay(150) #Hacer mas lento la mezcla

    inicia = True
    while True:
        event = pygame.event.wait()
        if event.type == pygame.QUIT:
            quit()
        elif event.type == pygame.MOUSEBUTTONDOWN:
            if inicia:
                ordena_piezas()
                inicia = False
            else:
                posicion_mouse = pygame.mouse.get_pos() #Obtener posicion del mouse (X,Y)
                r = int(posicion_mouse[0] / PIEZA_HEIGHT)
                c = int(posicion_mouse[1] / PIEZA_WIDTH)        
                intercambiar_piezas(r,c)
                if (posicion_inicial == posicion):                                
                    imagen = pygame.image.load(IMAGE_FILE)
                    pygame.init()
                    display = pygame.display.set_mode(IMAGE_SIZE)
                    pygame.display.set_caption("Rompecabezas")
                    display.blit (imagen, (0, 0))
                    pygame.display.flip() #Actualizar pantalla 
                    font = pygame.font.SysFont("arial",100)
                    fuente = font.render("¡Lo lograste!",True, (255,255,255))
                    texto = fuente.get_rect()
                    texto.center = (int(IMAGE_SIZE[0]/2),int(IMAGE_SIZE[1]/2))
                    display.blit(fuente, texto) 
                    pygame.display.update()

juego_rompecabezas()
```
